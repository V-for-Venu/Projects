## Data Analysis with Python

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import scipy.stats as stats
import warnings
import pandas as pd
warnings.filterwarnings("ignore")
```

```python
menuframe = pd.read_sql("SELECT * FROM MENU",mydb)
menuframe
```

```python
menuframe.describe(include = 'all')
```

```python
def plotswarm(X,Y,Data,axes):
    x= sns.swarmplot(x = X, y=Y, data = Data, ax = axes)
    plt.setp(x.get_xticklabels(), rotation = 65)
    axes.set_title(f"{Y} Content in the Menu", fontweight="bold")
    return x
```

```python
plots = ['Sodium', 'Total Fat','Cholesterol','Sugars']
fig, axes = plt.subplots(2,2, figsize = (12,9))
x = 'Category'

plotswarm(x,plots[0],menuframe,axes[0][0])  
plotswarm(x,plots[1],menuframe,axes[0][1]) 
plotswarm(x,plots[2],menuframe,axes[1][0]) 
plotswarm(x,plots[3],menuframe,axes[1][1]) 

plt.tight_layout()
plt.show()
```
