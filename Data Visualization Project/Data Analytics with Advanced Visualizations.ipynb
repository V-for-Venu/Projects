```python
canada_data_scatter = pd.DataFrame(canada_data_years[years].sum(axis = 0))
canada_data_scatter.reset_index(inplace = True)
canada_data_scatter
```

```python
years = np.arange(1980,2014)
#changing the datatype of years from str to integer
canada_data_scatter.columns = ['Year','Total']
canada_data_scatter['Year'] = years
```

```python
canada_data_scatter.plot(kind = 'scatter', x='Year' , y='Total', figsize = (10,6), color ='firebrick', s = 25)
#plt.xticks(np.arange(0,34,5))
plt.xlabel('Years')
plt.ylabel('Immigrants per Year')
plt.show()
```

```python
x = canada_data_scatter['Year']
y = canada_data_scatter['Total']
fitting_line = np.polyfit(x,y,deg = 1)
fitting_line
```

```python
#Plotting line on the scatter plot
canada_data_scatter.plot(kind = 'scatter', x='Year' , y='Total', figsize = (10,6), color ='firebrick', s = 25)

plt.xlabel('Years')
plt.ylabel('Immigrants per Year')
plt.title('TOTAL IMMIGRANTS TO CANADA FROM 1980 - 2013')
#plt.xticks(np.arange(0,34,3))
plt.plot(years, years * fitting_line[0] + fitting_line[1], color = 'deepskyblue')
plt.legend(['Regression Fitting Line','Data Points'])
plt.show()
```

## Advanced Visualizations

### Waffle Charts

```python
#Create new data with Norway, Swedan and Denmark
data_Scand = canada_data.loc[['Norway','Sweden','Denmark']].copy()
data_Scand
```

```python
width = 40
height = 15

total_tiles = width*height
print(f"Total Number of Ties : {total_tiles}")
```

```python
#We need to create our own waffles
total = data_Scand['Country Total'].sum()
proportions = data_Scand['Country Total']/total
proportions = pd.DataFrame({"Country Proportion":proportions})
proportions
```

```python
proportions['Tiles Count'] = (proportions['Country Proportion'] * total_tiles).round().astype(int)
proportions
```

```python
tiles_per_category = (proportions['Country Proportion'] * total_tiles).round().astype(int)
tiles_per_category = pd.DataFrame({"number of tiles":tiles_per_category})
tiles_per_category
```

```python
labels = list(proportions.index)
labels
```

```python
from pywaffle import Waffle
fig = plt.figure(FigureClass=Waffle, columns = 28, values = proportions['Tiles Count'], labels = labels,colors = ['red','lightgrey','blue'],legend = {'loc':(1,0.5)})
plt.tight_layout()
```

### Wordclouds

```python
from wordcloud import WordCloud, STOPWORDS
```

```python
total = canada_data['Country Total'].sum()
total
```

```python
word = ' '
max_words = 90
for i in canada_data.index:
    if i.count(" ")<2:
        rep = int(canada_data.loc[i,"Country Total"]/total*max_words)
        word = word + ((i + ' ') * rep )
print(word)
```

```python
wordcloud = WordCloud(background_color='white',width=1000, height=700).generate(word)
```

```python
plt.figure(figsize=(14, 18))

plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
